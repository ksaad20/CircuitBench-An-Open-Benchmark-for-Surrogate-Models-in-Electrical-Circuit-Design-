# Battery Management Benchmark Demo

This notebook demonstrates how to load, run, and evaluate the Battery Management benchmark in CircuitBench.

```python
from pathlib import Path
import json
```

```python
try:
    from circuitbench.benchmark import BenchmarkRunner
except ImportError:
    print("CircuitBench package not installed.")
```

```python
benchmark_dir = Path("../")
print("Benchmark directory:", benchmark_dir.resolve())
```

```python
results = {
    "benchmark": "battery_management",
    "status": "demo",
    "accuracy": None,
    "rmse": None,
    "mae": None,
}

print(results)
```

```python
output_dir = Path("outputs")
output_dir.mkdir(exist_ok=True)

with open(output_dir / "benchmark_summary.json", "w") as f:
    json.dump(results, f, indent=4)

print("Summary written to", output_dir / "benchmark_summary.json")
```
