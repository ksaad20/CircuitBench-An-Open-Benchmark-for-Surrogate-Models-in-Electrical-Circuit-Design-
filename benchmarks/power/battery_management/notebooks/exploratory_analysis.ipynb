# Exploratory Data Analysis

This notebook performs a basic exploratory analysis for the Battery Management benchmark.

```python
import json
from pathlib import Path

import numpy as np
import pandas as pd
```

```python
dataset_path = Path("../data/battery_management.csv")

if dataset_path.exists():
    df = pd.read_csv(dataset_path)
else:
    df = pd.DataFrame(
        {
            "Voltage": [3.65, 3.72, 3.81, 3.95, 4.05],
            "Current": [2.1, 1.8, 1.5, 1.1, 0.8],
            "Temperature": [25, 27, 29, 31, 34],
            "SOC": [18, 35, 54, 76, 95],
        }
    )

df.head()
```

```python
summary = df.describe(include="all")
summary
```

```python
missing_values = df.isna().sum()
missing_values
```

```python
correlation = df.select_dtypes(include=[np.number]).corr()
correlation
```

```python
report = {
    "rows": int(df.shape[0]),
    "columns": int(df.shape[1]),
    "features": list(df.columns),
    "missing_values": missing_values.to_dict(),
}

output_dir = Path("outputs")
output_dir.mkdir(exist_ok=True)

with open(output_dir / "eda_summary.json", "w") as f:
    json.dump(report, f, indent=4)

report
```
