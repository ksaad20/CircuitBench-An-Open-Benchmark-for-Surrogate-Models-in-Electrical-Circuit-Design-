# Converter Efficiency Benchmark Demo

This notebook demonstrates how to load the converter efficiency benchmark, inspect the dataset, train a baseline model, and evaluate its performance using CircuitBench.

```python
from pathlib import Path
import pandas as pd

print("Converter Efficiency Benchmark Demo")
```

```python
benchmark_dir = Path("../")
data_dir = benchmark_dir / "data"

print("Benchmark directory:", benchmark_dir.resolve())
print("Data directory:", data_dir.resolve())
```

```python
csv_files = list(data_dir.glob("*.csv"))

if csv_files:
    df = pd.read_csv(csv_files[0])
    display(df.head())
    print(df.describe(include="all"))
else:
    print("No CSV dataset found. Add a dataset to the data directory.")
```
