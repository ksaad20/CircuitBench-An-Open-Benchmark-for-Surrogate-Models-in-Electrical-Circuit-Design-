# Converter Efficiency - Exploratory Data Analysis

This notebook performs exploratory analysis of the converter efficiency benchmark dataset.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_dir = Path('../data')
csv_files = sorted(data_dir.glob('*.csv'))

if not csv_files:
    raise FileNotFoundError('No CSV dataset found in the data directory.')

df = pd.read_csv(csv_files[0])
print(df.shape)
display(df.head())
```

```python
display(df.describe(include='all'))
```

```python
numeric_df = df.select_dtypes(include='number')

if not numeric_df.empty:
    numeric_df.hist(figsize=(12, 8))
    plt.tight_layout()
    plt.show()
else:
    print('No numeric columns available.')
```

```python
if len(numeric_df.columns) > 1:
    corr = numeric_df.corr(numeric_only=True)
    display(corr)

    plt.figure(figsize=(10, 8))
    plt.imshow(corr, aspect='auto')
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.tight_layout()
    plt.show()
```
