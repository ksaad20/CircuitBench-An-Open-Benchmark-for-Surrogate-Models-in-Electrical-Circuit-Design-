# Thermal Prediction Benchmark Demo

This notebook demonstrates the Thermal Prediction benchmark included in CircuitBench.

```python
from pathlib import Path
import json
import pandas as pd
```

```python
benchmark_dir = Path("../")
data_dir = benchmark_dir / "data"

print(f"Benchmark directory: {benchmark_dir.resolve()}")
print(f"Data directory: {data_dir.resolve()}")
```

```python
csv_files = sorted(data_dir.glob("*.csv"))

if csv_files:
    df = pd.read_csv(csv_files[0])
    display(df.head())
    print(df.describe(include="all"))
else:
    df = pd.DataFrame(
        {
            "AmbientTemperature": [25, 27, 30, 32, 35],
            "PowerLoss": [10.2, 11.8, 13.1, 15.0, 17.3],
            "DeviceTemperature": [42.5, 45.8, 50.4, 55.2, 60.1],
        }
    )
    display(df)
```

```python
summary = {
    "benchmark": "thermal_prediction",
    "samples": int(df.shape[0]),
    "features": list(df.columns),
}

output_dir = Path("outputs")
output_dir.mkdir(exist_ok=True)

with open(output_dir / "benchmark_summary.json", "w") as file:
    json.dump(summary, file, indent=4)

summary
```
