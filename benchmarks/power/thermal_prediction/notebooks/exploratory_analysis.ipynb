# Thermal Prediction - Exploratory Data Analysis

This notebook performs exploratory data analysis for the Thermal Prediction benchmark.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_dir = Path("../data")
csv_files = sorted(data_dir.glob("*.csv"))

if csv_files:
    df = pd.read_csv(csv_files[0])
else:
    df = pd.DataFrame(
        {
            "AmbientTemperature": [20, 25, 30, 35, 40],
            "PowerLoss": [5.1, 7.4, 9.8, 12.2, 15.6],
            "Airflow": [1.2, 1.5, 1.8, 2.0, 2.4],
            "DeviceTemperature": [35.5, 40.8, 47.2, 54.1, 62.7],
        }
    )

display(df.head())
```

```python
print("Dataset shape:", df.shape)
display(df.describe(include="all"))
```

```python
missing_values = df.isna().sum()
display(missing_values)
```

```python
numeric_df = df.select_dtypes(include="number")

numeric_df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
correlation = numeric_df.corr(numeric_only=True)
display(correlation)

plt.figure(figsize=(8, 6))
plt.imshow(correlation, aspect="auto")
plt.xticks(
    range(len(correlation.columns)), correlation.columns, rotation=45, ha="right"
)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.colorbar()
plt.tight_layout()
plt.show()
```
