# CMOS Inverter Example

CircuitBench synthetic CMOS inverter benchmark dataset.

```python
import numpy as np
import pandas as pd

VDD = 1.8
Vin = np.linspace(0, VDD, 200)

Vout = VDD / (1 + np.exp(20 * (Vin - 0.9)))
Gain = np.gradient(Vout, Vin)

df = pd.DataFrame({
    "Input_Voltage": Vin,
    "Output_Voltage": Vout,
    "Voltage_Gain": Gain,
    "Technology": "CMOS_Inverter"
})

df.to_csv(
    "cmos_inverter_dataset.csv",
    index=False
)

df.head()
```
