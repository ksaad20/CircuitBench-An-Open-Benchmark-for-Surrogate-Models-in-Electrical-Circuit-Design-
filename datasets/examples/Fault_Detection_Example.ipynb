# Fault Detection Example

CircuitBench synthetic fault detection benchmark dataset.

```python
import numpy as np
import pandas as pd

np.random.seed(42)

samples = 1000

voltage = np.random.normal(5.0, 0.2, samples)
current = np.random.normal(2.0, 0.1, samples)
temperature = np.random.normal(40.0, 5.0, samples)

fault = np.random.choice(
    ["Normal", "Short_Circuit", "Open_Circuit"],
    size=samples,
    p=[0.80, 0.10, 0.10]
)

df = pd.DataFrame({
    "Voltage": voltage,
    "Current": current,
    "Temperature": temperature,
    "Fault_Class": fault
})

df.to_csv(
    "fault_detection_dataset.csv",
    index=False
)

df.head()
```
