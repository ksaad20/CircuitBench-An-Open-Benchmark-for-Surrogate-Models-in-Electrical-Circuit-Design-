# CircuitBench Dataset Example

This notebook demonstrates how to load and inspect an example dataset included with CircuitBench.

```python
from pathlib import Path

import pandas as pd
```

```python
dataset_dir = Path('../')
csv_files = sorted(dataset_dir.glob('*.csv'))

if csv_files:
    dataset = pd.read_csv(csv_files[0])
    print(f'Loaded: {csv_files[0].name}')
else:
    dataset = pd.DataFrame(
        {
            'Voltage': [3.30, 3.45, 3.60, 3.75, 3.90],
            'Current': [0.50, 0.65, 0.80, 0.95, 1.10],
            'Power': [1.65, 2.24, 2.88, 3.56, 4.29]
        }
    )
    print('No CSV file found. Using example dataset.')

display(dataset.head())
```

```python
print('Dataset Shape:', dataset.shape)
display(dataset.describe(include='all'))
```

```python
print('Column Names:')
for column in dataset.columns:
    print('-', column)
```
