# Leaderboard Analysis

## CircuitBench Tutorial 10

This notebook demonstrates comprehensive analysis of benchmark leaderboards generated by CircuitBench. It ranks machine learning models, compares statistical performance, identifies Pareto-optimal solutions, and generates publication-quality tables and visualizations.

### Learning Objectives

- Load benchmark leaderboard results
- Rank competing models
- Compare evaluation metrics
- Identify Pareto-optimal models
- Visualize leaderboard performance
- Export publication-ready reports


## Scientific Background

Benchmark leaderboards summarize the comparative performance of machine learning models using standardized evaluation metrics. Statistical ranking and reproducible reporting enable fair comparison across algorithms and benchmark datasets.

```python
import random
import warnings
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
```

```python
from circuitbench.analysis import LeaderboardAnalyzer
```

## Load Leaderboard

```python
leaderboard = pd.read_csv("baseline_results/leaderboard.csv")

leaderboard.head()
```

## Initialize Analyzer

```python
analyzer = LeaderboardAnalyzer()
```

## Leaderboard Overview

```python
print("Models:", len(leaderboard))
print("Columns:", list(leaderboard.columns))

leaderboard.describe(include="all")
```

## Rank Models

```python
leaderboard = leaderboard.sort_values(by="RMSE", ascending=True).reset_index(drop=True)

leaderboard["Rank"] = np.arange(1, len(leaderboard) + 1)

leaderboard
```

## Top Performing Models

```python
top_models = leaderboard.head(10)

top_models
```

## Summary Statistics

```python
statistics = leaderboard.describe()

statistics
```

## Best Model

```python
best_model = leaderboard.iloc[0]

best_model
```

## RMSE Ranking

```python
plt.figure(figsize=(10, 6))
plt.bar(leaderboard["Model"], leaderboard["RMSE"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("RMSE")
plt.title("Model Ranking by RMSE")
plt.tight_layout()
```

## MAE Ranking

```python
plt.figure(figsize=(10, 6))
plt.bar(leaderboard["Model"], leaderboard["MAE"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("MAE")
plt.title("Model Ranking by MAE")
plt.tight_layout()
```

## R² Comparison

```python
plt.figure(figsize=(10, 6))
plt.bar(leaderboard["Model"], leaderboard["R2"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("R² Score")
plt.title("Coefficient of Determination")
plt.tight_layout()
```

## Export Ranked Leaderboard

```python
output_dir = Path("leaderboard_analysis")
output_dir.mkdir(exist_ok=True)

leaderboard.to_csv(output_dir / "ranked_leaderboard.csv", index=False)
```

## Pareto Frontier Analysis

```python
leaderboard["Pareto_Optimal"] = analyzer.compute_pareto_front(
    leaderboard, objectives={"RMSE": "min", "MAE": "min", "R2": "max"}
)

leaderboard[["Model", "Pareto_Optimal"]]
```

## Pareto Visualization

```python
plt.figure(figsize=(8, 6))
plt.scatter(leaderboard["RMSE"], leaderboard["R2"])
plt.xlabel("RMSE")
plt.ylabel("R²")
plt.title("Pareto Analysis")
plt.tight_layout()
```

## Model Performance Table

```python
performance_table = leaderboard[
    ["Rank", "Model", "MAE", "RMSE", "R2", "Pareto_Optimal"]
]

performance_table
```

## Statistical Summary

```python
summary = {
    "Number of Models": int(len(leaderboard)),
    "Best RMSE": float(leaderboard["RMSE"].min()),
    "Worst RMSE": float(leaderboard["RMSE"].max()),
    "Mean RMSE": float(leaderboard["RMSE"].mean()),
    "Median RMSE": float(leaderboard["RMSE"].median()),
}

pd.Series(summary)
```

## Export Analysis Tables

```python
performance_table.to_csv(output_dir / "performance_table.csv", index=False)

statistics.to_csv(output_dir / "leaderboard_statistics.csv")
```

## Export Leaderboard Report

```python
import json

report = {
    "best_model": str(best_model["Model"]),
    "best_rmse": float(best_model["RMSE"]),
    "best_mae": float(best_model["MAE"]),
    "best_r2": float(best_model["R2"]),
    "number_of_models": int(len(leaderboard)),
    "pareto_models": int(leaderboard["Pareto_Optimal"].sum()),
    "random_seed": SEED,
}

with open(output_dir / "leaderboard_report.json", "w") as f:
    json.dump(report, f, indent=4)
```

## Best Practices

1. Evaluate every model using identical train/test splits.
2. Report multiple evaluation metrics instead of relying on a single score.
3. Include computational cost alongside predictive performance.
4. Archive leaderboard versions to ensure reproducibility.
5. Highlight Pareto-optimal models instead of only the top-ranked model.
6. Report confidence intervals when repeated experiments are available.


## Publication Recommendations

For manuscripts, include:
