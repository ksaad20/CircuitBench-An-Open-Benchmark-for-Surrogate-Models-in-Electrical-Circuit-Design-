# CircuitBench

## An Open Benchmark for Surrogate Models in Electrical Circuit Design

### Getting Started

This notebook demonstrates how to:

- Load CircuitBench
- Explore benchmark datasets
- Train a baseline model
- Evaluate performance
- Visualize results
- Export experiment reports


```python
from circuitbench.benchmark import BenchmarkRunner

runner = BenchmarkRunner()
runner.available_benchmarks()
```

```python
benchmark = runner.load(benchmark_name="CMOS_Inverter")
benchmark.summary()
```

```python
X_train, X_test, y_train, y_test = benchmark.train_test_split(
    test_size=0.2, random_state=42
)
```

```python
model = runner.create_model("RandomForest")
predictions = model.predict(X_test)
results = runner.evaluate(y_test, predictions)
results
```
